It is interesting to note that in the embeddings of size 100 (with and without dropout) a lot of entries are 0.



```python
# make predictions
pred_file = "../handout/"+'/predictions_mobilenet_emb100_batch30_epoch5_dropout_img224x224.txt'

emb_A = np.loadtxt("../handout/"+'/emb_A_mobilenet_emb100_batch30_epoch5_dropout_img224x224.txt')
emb_B = np.loadtxt("../handout/"+'/emb_B_mobilenet_emb100_batch30_epoch5_dropout_img224x224.txt')
emb_C = np.loadtxt("../handout/"+'/emb_C_mobilenet_emb100_batch30_epoch5_dropout_img224x224.txt')

predictions = np.zeros(len(emb_A))
for i in range(len(emb_A)):
    dist_AB = math.pow(2,np.linalg.norm(emb_A[i,:] - emb_B[i,:]))
    dist_AC = math.pow(2,np.linalg.norm(emb_A[i,:] - emb_C[i,:]))

    if dist_AB < dist_AC:
        predictions[i] = 1 
        continue
    #then dist_AB >= dist_AC
    predictions[i] = 0

np.savetxt(pred_file, predictions, fmt='%d')
```
